### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
        "AddressLine3":"London", "RatingValue": 
        {"$gte":4}
        }
                                                  
result = establishments.find(query)
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f" The number of establishments is {count}.")

# Display the first document in the results using pprint
establishments.find_one(query)
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(list(result))
# Display the number of rows in the DataFrame
count = len(london_df)
print(f" The number of rows in london_df is {count}.")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#Display Penang Flavours document 

establishments.find_one({'BusinessName':'Penang Flavours'})
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

min_lat = latitude - degree_search
max_lat = latitude + degree_search
min_lon = longitude - degree_search
max_lon = longitude + degree_search

print("Latitude range:", min_lat, max_lat)
print("Longitude range:", min_lon, max_lon)

query = {"RatingValue": 5,
         'geocode.latitude': 
         {"$gte": min_lat, "$lte": max_lat},
         'geocode.longitude': 
         {"$gte": min_lon, "$lte": max_lon}
        }

sort = [('scores.Hygiene',1)]
limit = 5

# Print the results
pprint(list(establishments.find(query).sort(sort).limit(limit)))
```

```python
# Convert result to Pandas DataFrame
df_1 = pd.DataFrame(list(establishments.find(query).sort(sort).limit(limit)))
df_1.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Define the aggregation pipeline

pipeline = [
    {
        "$match": 
        {"scores.Hygiene": 0}
    },
    {
        "$group": 
        {"_id": "$LocalAuthorityName", "count": {"$sum": 1}
        }
    },
    {
        "$sort": 
        {"count": -1}
    },
]

# Execute the aggregation query
results = establishments.aggregate(pipeline)

# Store results in a list to avoid exhausting the cursor
results_list = list(results)

# Print the number of documents in the result
print(f"The number of documents in the list is {len(results_list)}.")

# Print the first 10 results
print("First 10 results:")
for result in results_list[:10]:  
    print(result)
```

```python
# Convert the result to a Pandas DataFrame
pd_2 = pd.DataFrame(results_list)
# Display the number of rows in the DataFrame
print(f" The number of rows in the DataFrame is {len(pd_2)}.")
# Display the first 10 rows of the DataFrame
pd_2.head(10)
```
